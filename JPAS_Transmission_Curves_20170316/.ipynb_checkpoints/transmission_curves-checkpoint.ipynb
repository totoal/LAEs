```python
import matplotlib.pyplot as plt
from os import listdir
import numpy as np
```

```python
filters_tags = []
for item in listdir():
    if item.find('.tab') != -1:
        filters_tags.append(item)
```

```python
# Load transmission curves in a matrix

tcurves = [[*range(3000,11000)]]

for tag in filters_tags:
    
    f = open(tag, mode='r')
    lines = f.readlines()[12:]
    column = []
    
    for l in range(len(lines)):
        column.append(float(lines[l].split()[1]))
    
    tcurves.append(column)

tcurves = np.array(tcurves).transpose()
```
